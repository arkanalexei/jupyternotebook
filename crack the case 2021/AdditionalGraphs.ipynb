```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = df = pd.ExcelFile('data - Copy.xlsx', engine='openpyxl')
pillar = pd.read_excel(df, 'Pillars')
cols = [
    'isonum',
    'isocode',
    'PI',
    'econ',
    'busi',
    'gove',
    'educ',
    'heal',
    'safe',
    'pers',
    'soci',
    'envi'
]
pillar.drop(cols, axis=1, inplace=True)
pillar
```

```python
pillar.columns
```

```python
kenya = pillar.loc[pillar['country'] == 'Kenya']
malawi = pillar.loc[pillar['country'] == 'Malawi']
macedonia = pillar.loc[pillar['country'] == 'Macedonia']
kuwait = pillar.loc[pillar['country'] == 'Kuwait']
```

```python
country = kuwait

fig, axs = plt.subplots(3,3, figsize=(20,20))

plt.setp(axs, xticks=country['year'], xticklabels=country['year'])

axs[0, 0].plot(country['year'], country['rank_econ'], 'tab:orange')
axs[0, 0].set_title('Economic Quality', fontsize='xx-large')

axs[0, 1].plot(country['year'], country['rank_busi'], 'tab:green')
axs[0, 1].set_title('Business', fontsize='xx-large')

axs[0, 2].plot(country['year'], country['rank_gove'], 'tab:red')
axs[0, 2].set_title('Governance', fontsize='xx-large')

axs[1, 0].plot(country['year'], country['rank_educ'], 'tab:purple')
axs[1, 0].set_title('Education', fontsize='xx-large')

axs[1, 1].plot(country['year'], country['rank_heal'], 'tab:brown')
axs[1, 1].set_title('Health', fontsize='xx-large')

axs[1, 2].plot(country['year'], country['rank_safe'], 'tab:pink')
axs[1, 2].set_title('Safety and Security', fontsize='xx-large')

axs[2, 0].plot(country['year'], country['rank_pers'], 'tab:blue')
axs[2, 0].set_title('Personal Freedom', fontsize='xx-large')

axs[2, 1].plot(country['year'], country['rank_soci'], 'tab:olive')
axs[2, 1].set_title('Social Capital', fontsize='xx-large')

axs[2, 2].plot(country['year'], country['rank_envi'], 'tab:cyan')
axs[2, 2].set_title('Natural Environment', fontsize='xx-large')

mydict = {
    (0,0): "rank_econ",
    (0,1): "rank_busi",
    (0,2): "rank_gove",
    (1,0): "rank_educ",
    (1,1): "rank_heal",
    (1,2): "rank_safe",
    (2,0): "rank_pers",
    (2,1): "rank_soci",
    (2,2): "rank_envi",
}

for cord, col in mydict.items():
    for a,b in zip(country['year'], country[col]):
        axs[cord].text(a, b, str(b), fontsize='large')
        
for cord in mydict.keys():
    axs[cord].invert_yaxis()
    
for cord in mydict.keys():
    axs[cord].axvspan(2012,2013, alpha=0.1, color='red')
```
