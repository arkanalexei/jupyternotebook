```python
# Import necessary modules
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
# Removing unecessary columns
data = pd.read_excel('positivity rate.xlsx', engine='openpyxl')
cols = [
    'Unnamed: 9',
    'Unnamed: 10',
    'Unnamed: 11',
    'Unnamed: 12',
    'Unnamed: 13',
    'Unnamed: 14',
    'Unnamed: 15',
    'Unnamed: 16',
    'Unnamed: 17',
    'Unnamed: 18',
    'Unnamed: 19',

]

data.drop(cols, inplace=True, axis=1)
data.drop(list(range(502,867)), inplace=True, axis=0)
```

```python
# Using data from April 2020 onwards
data2 = data[data.Tanggal >= '2020-04-2']
```

```python
x = data2['Tanggal']
y = data2['Positivity Rate Kasus Baru Harian'] * 100


ticks = [
    '2020-04-01',
    '2020-05-01',
    '2020-06-01',
    '2020-07-01',
    '2020-08-01',
    '2020-09-01',
    '2020-10-01',
    '2020-11-01',
    '2020-12-01',
    '2021-01-01',
    '2021-02-01',
    '2021-03-01',
    '2021-04-01',
    '2021-05-01',
    '2021-06-01',
    '2021-07-01',
    
]

figure(figsize=(16,10))
plt.plot(x,y, lw=2.5, color='navy')
plt.xticks(ticks, rotation=90)

# Highlight 
plt.axvspan('2020-04-10','2020-06-04', color='orange', alpha=0.3, label='PSBB Jilid I-III')
plt.axvspan('2020-06-05','2020-09-13', color='green', alpha=0.3, label='PSBB Transisi')
plt.axvspan('2020-09-14','2020-10-11', color='orange', alpha=0.3, label='PSBB Ketat')
plt.axvspan('2020-10-12','2021-01-10', color='green', alpha=0.3, label='PSBB Transisi II')
plt.axvspan('2021-01-11','2021-07-02', color='red', alpha=0.3, label='PPKM')
plt.axvspan('2021-07-03','2021-07-14', color='darkred', alpha=0.3, label='PPKM Darurat')
plt.axvspan('2021-05-03','2021-05-03', color='red', alpha=1, label='Varian Delta Teridentifikasi Pertama Kali')
plt.legend(loc='upper left')

plt.title('Positivity Rate Kasus Baru Harian')
plt.show()
```

Grafik dibawah menunjukkan bahwa jumlah orang positif harian itu sejajar dengan jumlah orang di tes. Hal ini karena semakin banyak orang yang di tes, kemungkinan orang yang terdeteksi positif juga meningkat.

Dan sekali lagi, kita bisa melihat bahwa efektivitas PSBB Transisi jauh kalah dengan PPKM untuk mengurangi mobilitas.
